```python
print("df_1:", df_1.columns.tolist())
print("df_hoteles:", df_hoteles.columns.tolist())
print("df_vuelos:", df_vuelos.columns.tolist())
```

```python

# Merge con hoteles (many-to-many)
df_1_hoteles = pd.merge(df_1, df_hoteles, how='left', left_on='ciudad', right_on='destination_city')

# Merge con vuelos (many-to-many)
total_data = pd.merge(df_1_hoteles, df_vuelos, how='left', left_on='ciudad', right_on='destination_city')

# Tomar una muestra de 240,000 filas (o menos si no hay tantas)
sample_size = min(240_000, len(total_data))
sample_data = total_data.sample(n=sample_size, random_state=42)

# Guardar resultados
sample_data.to_csv("total_data_240k.csv", index=False)
sample_data.to_csv("total_data_240k.csv.gz", index=False, compression="gzip")

print("✅ Archivos guardados como total_data_240k.csv y total_data_240k.csv.gz")
```
